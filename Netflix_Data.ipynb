# 1) Find out which types of genre has most movies and TV Shows

```python
import pandas as pd

file_path =('/content/netflix_movies .csv.csv')
df = pd.read_csv(file_path)
df['listed_in'] = df['listed_in'].fillna('Unknown')
genres = df['listed_in'].str.split(', ', expand=True).stack().value_counts()

print("Top genres by count:")
print(genres.head())
```

 2) What is the distribution of content types (e.g., Movies vs. TV Shows) in the dataset?


```python
content_distribution = df['type'].value_counts()

print("Distribution of content types:")
print(content_distribution)
```

3) Which countries contribute the most to the content library?


```python
df['country'] = df['country'].fillna('Unknown')
countries = df['country'].str.split(', ', expand=True).stack().value_counts()

print("\nTop contributing countries:")
print(countries.head())
```

5) What are the most common ratings, and how are they distributed across different types of content?



```python
df['rating'] = df['rating'].fillna('Unknown')
rating_distribution = df['rating'].value_counts()

print("\nMost common ratings:")
print(rating_distribution)
```
